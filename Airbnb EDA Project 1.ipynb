### `Visualization of listing counts distribution of the rooms and average price according to the room type`

```python
fig, axes = plt.subplots(1,3,figsize=(15,5))
axes[0].bar(listing_count.index,listing_count.values,color = 'skyblue')
axes[1].pie(listing_percent,labels = listing_percent.index,autopct='%1.1f%%',colors=['lightcoral', 'lightgreen', 'lightblue'])
axes[2].bar(Roomtype_avgprice.index,Roomtype_avgprice.values,color='lightgreen')


# Function to add value labels
def add_value_labels(ax, spacing=5):
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label.
        space = spacing
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:.1f}".format(y_value)

        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(0, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for positive and negative values

# Call the function for each subplot
add_value_labels(axes[0])
add_value_labels(axes[2])

plt.tight_layout()

plt.show()
```

### `Comparison of Manhattan's average price to the other's neighborhood areas or location also comparison of listing counts with the Manhattan with the others location`

```python
import matplotlib.pyplot as plt

# Plotting average prices
plt.figure(figsize=(10, 5))
plt.barh(list(average_prices_comparison.keys()), average_prices_comparison.values(), color='skyblue')
plt.axvline(x=avg_Price_MANHATTAN, color='red', linestyle='--', label='Manharrab Avg Price')
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.title('Comparison of Average Prices with Manharrab')
plt.legend()
plt.tight_layout()

# add label to bar
for index, value in enumerate(average_prices_comparison.values()):
    plt.text(value, index, str(round(value, 2)), color='black', ha='left', va='center')

plt.tight_layout()
plt.show()

# Plotting listing counts
plt.figure(figsize=(10, 5))
plt.barh(list(listing_counts_comparison.keys()), listing_counts_comparison.values(), color='lightgreen')
plt.axvline(x=manhanttan_listing_count, color='red', linestyle='--', label='Manharrab Listing Count')
plt.xlabel('Listing Count')
plt.ylabel('Neighborhood')
plt.title('Comparison of Listing Counts with Manharrab')
plt.legend()
plt.tight_layout()

# add labels to the graph
for index, value in enumerate(listing_counts_comparison.values()):
    plt.text(value, index, str(value), color='black', ha='left', va='center')

plt.tight_layout()
plt.show()
```

### `Visualization Host who got Maxim number of reviews according to the location`

```python
# Reset the index to make it easier to access 'host_name' and 'neighbourhood_group' for plotting
host_review_top5 = host_review_top5.reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(host_review_top5)), host_review_top5['number_of_reviews'], color='skyblue')

# labels to bar chart
for bar, label in zip(bars, host_review_top5['number_of_reviews']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, label, ha='center', va='bottom')

plt.xlabel('Host and Neighbourhood Group')
plt.ylabel('Maximum Number of Reviews')
plt.title('Top 5 Host and Neighbourhood Group Combinations with Maximum Number of Reviews')
plt.xticks(range(len(host_review_top5)), [f"{name}, {neighbourhood_group}" for name, neighbourhood_group in zip(host_review_top5['host_name'], host_review_top5['neighbourhood_group'])], rotation=45, ha='right')
plt.tight_layout()


plt.show()
```
